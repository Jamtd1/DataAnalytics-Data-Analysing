### Check for duplicate rows/columns

Two ways to check for the presence of duplicate rows are used below.  

The first method ignores the first occurance of a duplicate row, i.e. if line 200 was reported as a duplicate then it is a duplicate of at least one line between 0 - 199.  

The second method counts every duplicate line it comes accross in the data frame. It will keep the first occurance of the duplicate lines in the data too. Hence, this type will have more duplicate lines than the previous method.  


```python
# the first method, df.duplicated() assumes keep=First
print('Number of duplicate rows, excluding the first rows, in the table: ', df.duplicated().sum())

# the second method, not df.duplicated(keep=False)
print('Number of duplicaet rows, including the first rows, in the table: ', df.duplicated(keep=False).sum())
```

The first feature pair to investigate is the categorical-categorical plot of the age group and sex who are at a higher risk of contracting the virus. To complete this analysis a back-to-back bar plot will be used to highlight the age groups and which sex is more likely to contract the virus.

```python
# Take a slice of the main dataframe for use in this pairing
df_age_sex = df[['sex', 'age_group']]
```

```python
# Create a series containing the counts of females in each age bracket
female_true = df_age_sex.apply(lambda x: True if (x['sex'] == "Female") else False, axis=1)

female_age_list = female_true.tolist()

female_ages = df_age_sex['age_group'].iloc[female_age_list]
female_ages_count = female_ages.value_counts(sort=False)

#print(female_ages_count)
```

```python
# Create a series containing the counts of males in each age bracket
male_true = df_age_sex.apply(lambda x: True if (x['sex'] == "Male") else False, axis=1)

male_age_list = male_true.tolist()

male_ages = df_age_sex['age_group'].iloc[male_age_list]
male_ages_count = male_ages.value_counts(sort=False)

#print(male_ages_count)
```

```python
# Create a list of just the counts
female_counts_dict = female_ages_count.to_dict()
#print(counts_dict)
female_counts_list = []

for i in female_counts_dict:
    female_counts_list.append(female_counts_dict[i])

# used to remove the zero for missing
del female_counts_list[-1]

#print(female_counts_list)
```

```python
# Create a list of just the counts
male_counts_dict = male_ages_count.to_dict()
#print(counts_dict)
male_counts_list = []

for i in male_counts_dict:
    male_counts_list.append(male_counts_dict[i])

#used to remove the zero for missing
del male_counts_list[-1]

#print(male_counts_list)
```

```python
# Create a list of the cardinalities
age_list = []

for i in female_counts_dict:
    age_list.append(i)
age_list.sort()

# used to remove missing
del age_list[-1]

#print(age_list)
```

```python
# Create a joined list of the male and female counts in a list of lists
fage_mage_lists = []
i = 0
while i < len(female_counts_list):
    fage_mage_lists.append([female_counts_list[i], male_counts_list[i]])
    i += 1

#print(fage_mage_lists)
```

```python
# Merge the age_list and the list of lists to make a dictionary
sex_vs_age_dict = {}
i = 0
for age in age_list:
    sex_vs_age_dict[age] = fage_mage_lists[i]
    i += 1

#print(sex_vs_age_dict)
```

```python
# Turn the dictionary into a new data frame
sex_vs_age_df = pd.DataFrame.from_dict(sex_vs_age_dict)
```

```python
sex_vs_age_df
```

```python
# assign row zero to x-axis 1 and row 1 to x-axis 2
x1 = sex_vs_age_df.iloc[0, 0:]
x2 = sex_vs_age_df.iloc[1, 0:]
```

```python
# assign the bar_labels
bar_labels = age_list
```

```python
fig = plt.figure(figsize=(8,6))
```

```python
y_pos = np.arange(len(x1))
y_pos = [x for x in y_pos]
plt.yticks(y_pos, bar_labels, fontsize=10)


plt.barh(y_pos, x1, align='center', alpha=0.4, color='#C800FF')

plt.barh(y_pos, -x2, align='center', alpha=0.4, color='#21CBFF')

plt.xlabel('Male: Light Blue. Female: Electric Purple')
t = plt.title('Comparison of Male and Female age groups with covid')
plt.ylim([-1, len(x1) +0.1])
plt.xlim([-max(x2)-10, max(x1)+10])
plt.grid()

plt.show()
```
